```python
import pandas as pd
Teams = {
    'cities': ['London', 'Madrid', 'Turin', 'Munich'],
    'clubs' : ['Chelsea', 'Real Madrid', 'Juventus', 'Bayern Munich']
}
TeamA= pd.DataFrame(Teams)
TeamA
```

```python
Sponsership = {
    'clubs' :['Chelsea','Real Madrid', 'Juventus','Manchester United'],
    'sponsers':['Yokohama', 'Fly Emirates', 'Jeep', 'Chervolet']
}
SponsershipB = pd.DataFrame(Sponsership)
SponsershipB
```

```python
TeamA.merge(SponsershipB, on = 'clubs', how= 'inner')//only takes keys and values from both frames
```

```python
TeamA.merge(SponsershipB, on = 'clubs', how= 'outer')// takes all keys and values from both frames, also provides a NaN values for objects without values
```

```python
TeamA.merge(SponsershipB, on = 'clubs', how= 'right') // takes keys and values  from right frame only 
```

```python
TeamA.merge(SponsershipB, on = 'clubs', how= 'left')// take keys and values from the left frame only 
```
